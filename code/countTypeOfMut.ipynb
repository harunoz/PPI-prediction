```python
data = pd.read_csv('finalLLLL.csv')
```

```python
mutation=0
mutationCasInter=0
mutationDecInter=0
mutationDecInterRate=0
mutationDecInterS=0
mutationDistInter=0
mutationDistInterRate=0
mutationDistInterS=0
mutationIncInter=0
mutationIncInterRate=0
mutationIncInterS=0
mutationNoEffect=0
error=0

for i in range(len(data)):
    if(data.iloc[i,1][-9:]=='(MI:0118)'):
        mutation=mutation+1
    elif(data.iloc[i,1][-9:]=='(MI:2227)'):
        mutationCasInter=mutationCasInter+1
    elif(data.iloc[i,1][-9:]=='(MI:0119)'):
        mutationDecInter=mutationDecInter+1
    elif(data.iloc[i,1][-9:]=='(MI:1130)'):
        mutationDecInterRate=mutationDecInterRate+1
    elif(data.iloc[i,1][-9:]=='(MI:1133)'):
        mutationDecInterS=mutationDecInterS+1
    elif(data.iloc[i,1][-9:]=='(MI:0573)'):
        mutationDistInter=mutationDistInter+1
    elif(data.iloc[i,1][-9:]=='(MI:1129)'):
        mutationDistInterRate=mutationDistInterRate+1
    elif(data.iloc[i,1][-9:]=='(MI:1128)'):
        mutationDistInterS=mutationDistInterS+1
    elif(data.iloc[i,1][-9:]=='(MI:0382)'):
        mutationIncInter=mutationIncInter+1
    elif(data.iloc[i,1][-9:]=='(MI:1131)'):
        mutationIncInterRate=mutationIncInterRate+1
    elif(data.iloc[i,1][-9:]=='(MI:1132)'):
        mutationIncInterS=mutationIncInterS+1
    elif(data.iloc[i,1][-9:]=='(MI:2226)'):
        mutationNoEffect=mutationNoEffect+1
    else:
        print(data.iloc[i,0],data.iloc[i,1])
        error=error+1
        

```

```python
print('Mutation (MI:0118)', mutation)
print('Mutation causing an interaction (MI:2227))', mutationCasInter)
print('Mutation decreasing interaction (MI:0119)', mutationDecInter)
print('Mutation decreasing interaction rate (MI:1130))', mutationDecInterRate)
print('Mutation decreasing interaction strength (MI:1133))', mutationDecInterS)
print('Mutation disrupting interaction (MI:0573))', mutationDistInter)
print('Mutation disrupting interaction rate (MI:1129))', mutationDistInterRate)
print('Mutation disrupting interaction strength (MI:1128)', mutationDistInterS)
print('Mutation increasing interaction (MI:0382)', mutationIncInter)
print('Mutation increasing interaction rate (MI:1131)',mutationIncInterRate)
print('Mutation increasing interaction strength (MI:1132)', mutationIncInterS)
print('Mutation with no effect (MI:2226)', mutationNoEffect)
```
